```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import io
from collections import Counter
import snscrape.modules.twitter as sntwitter
import re
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('max_colwidth', 40)
```

```python
fortune=pd.read_excel('fortune.xlsx')
l_aerospace=fortune.iloc[:,0].dropna().tolist()
l_computer_hw=fortune.iloc[:,1].dropna().tolist()
l_computer_sw=fortune.iloc[:,2].dropna().tolist()
l_apparel=fortune.iloc[:,3].dropna().tolist()
l_banks=fortune.iloc[:,4].dropna().tolist()
l_div_fin=fortune.iloc[:,5].dropna().tolist()
l_entertainment=fortune.iloc[:,6].dropna().tolist()
#l_fin_dataserv=fortune.iloc[:,7].dropna().tolist()
#l_food_consumer=fortune.iloc[:,8].dropna().tolist()
l_food_services=fortune.iloc[:,7].dropna().tolist()
l_it_services=fortune.iloc[:,8].dropna().tolist()
l_iservice_retail1=fortune.iloc[:,9].dropna().tolist()
```

```python
#l_aerospace
n_limit=999
tweets = []

for user in l_aerospace:
  for i,tweet in tqdm(enumerate(sntwitter.TwitterSearchScraper('from:' + user).get_items()), total=n_limit):
    if len(tweets)>n_limit:
      n_limit+=1000
      break
    tweets.append([tweet.user.username,    tweet.date,   tweet.rawContent,  tweet.user.followersCount, tweet.links,  tweet.replyCount,     tweet.quoteCount,     tweet.lang,
                   tweet.retweetedTweet,    tweet.quotedTweet,      tweet.mentionedUsers,        tweet.retweetCount,       tweet.likeCount              ])
  print('done ' + user)


aerospace = pd.DataFrame(tweets)

aerospace= aerospace.set_axis(['username',                'date',               'content',              'followerscount', 'link',      'replycount',    'quoteCount',
                'language',    'retweetedtweet',        'quotedtweet',           'mentionedusers',          'retweetcount',             'likecount'], axis=1)
aerospace.groupby("username").mean()
```
